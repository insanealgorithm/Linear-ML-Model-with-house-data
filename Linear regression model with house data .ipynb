```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
house = pd.read_csv('House.csv')
```

```python
house.head()
```

```python
house.info()
```

```python
house.mainroad.value_counts()
```

```python
house.mainroad = house.mainroad.replace(['yes','no'],['1','0'])
house
```

```python
house.guestroom = house.guestroom.replace(['yes','no'],['1','0'])
house
```

```python
house.basement = house.basement.replace(['yes','no'],['1','0'])
house
```

```python
house.hotwaterheating = house.hotwaterheating.replace(['yes','no'],['1','0'])
house.airconditioning = house.airconditioning.replace(['yes','no'],['1','0'])
house.prefarea = house.prefarea.replace(['yes','no'],['1','0'])
house
```

```python
house.furnishingstatus.value_counts()
```

```python
house.furnishingstatus = house.furnishingstatus.replace(['furnished','unfurnished','semi-furnished'],['1','0','2'])
```

```python
house
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = house.drop('price',axis=1)
```

```python
X
```

```python
y = house.price
```

```python
y
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.30)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
prediction = model.predict(X_test)
```

```python
prediction
```

```python
sns.scatterplot(y_test,prediction)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,prediction)
```
